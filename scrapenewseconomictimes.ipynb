```python
from bs4 import BeautifulSoup
import requests
url = "https://economictimes.indiatimes.com/tech/newsletters/ettech-unwrapped/edtech-ma-talks-gain-momentum-amid-a-prolonged-funding-crunch/articleshow/98221912.cms"
response = requests.get(url)

response
```

```python
soup = BeautifulSoup(response.text, "html.parser")
links = []
for link in soup.findAll('a'):
    links.append(link.get('href'))

links

```

```python
import requests
from bs4 import BeautifulSoup
for url in links:
# Send a request to the webpage
    #url = "https://economictimes.indiatimes.com/wealth/personal-finance-news/why-income-tax-basic-exemption-limit-should-be-increased-to-rs-5-lakh-in-budget-2023/articleshow/96761500.cms"
    response = requests.get(url)

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract the title, description, and content
    title = soup.find("title").text
    description = soup.find("meta", {"name": "description"})["content"]
    content = "\n".join([p.text for p in soup.find_all("div", {"class": "Normal"})])

    # Print the title, description, and content
    print("Title:", title)
    print("Description:", description)
    print("Content:", content)
```
